```python
import re
def get_aspects(text):
    # text = text.lower()

    aspect_detected = {}
    for idx,aspect in enumerate(wiki_aspect_list):
        #print(idx)
        if re.search(r"\b" + re.escape(aspect) + r"\b", text):
            aspect_count = len(re.findall(r"\b" + re.escape(aspect) + r"\b", text))
            # print(aspect_count)
            aspect_detected[aspect] = aspect_count

    weighted_aspect_dict = {k: v / total for total in (sum(aspect_detected.values()),)
                            for k, v in aspect_detected.items()}
    return weighted_aspect_dict
```

```python
aspects_list_args = []
for idx, text in enumerate(argsme_datasets_text_list):
    
    aspects = get_aspects(text)
    aspects_list_args.append(aspects)
    # print(idx, aspects)
```
